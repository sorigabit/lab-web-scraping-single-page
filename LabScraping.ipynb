```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 2. find url and store it in a variable
url = "https://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
# 3. download html with a get request
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# # 4.2. check that the html code looks like it should
# soup
```

```python
# 5. retrieve/extract the desired info (here, you'll paste the "Selector" you copied before to get the element that belongs to the top movie)
# soup.select("body > div.container > div:nth-child(4) > div.col-xs-12.col-md-8 > div.chart-wrapper")
```

```python
#  as it's my first scraping, i will do one by one and check the results each time
```

```python
artist=[]
for art in soup.select('.artist'):
    artist.append(art.get_text())
```

```python
artist
```

```python
song=[]
```

```python
for s in soup.select('.title'):
    song.append(s.get_text())
```

```python
song
```

```python
position=[]
for cp in soup.select('.chart-position'):
    position.append(cp.get_text())
```

```python
position
```

```python
chart = pd.DataFrame({"chart_position":position, "artist":artist, "song":song})
```

```python
chart
```
